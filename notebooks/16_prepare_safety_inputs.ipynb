```python
import pandas as pd

# Load the latest master table that includes CCTV
df = pd.read_csv("../data/mdm2_data_files/big_table_with_weather_rain_clusters_cctv300m.csv")

# Keep only sensor-level unique values
team = (
    df[["sensor_id", "cctv_300m"]]
    .drop_duplicates("sensor_id")
    .sort_values("sensor_id")
    .reset_index(drop=True)
)

OUT = "../reports/sensor_safety_inputs_template.csv"
team.to_csv(OUT, index=False)

print("Saved:", OUT)
print(team.head())
print("Rows (sensors):", len(team))
```
